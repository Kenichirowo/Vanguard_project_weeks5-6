# COMPLETION RATE

```python
# We can add an additional check to see whether a step is followed by a previous step, indicating an error (backtrack).
funnel_orders = ['start', 'step_1', 'step_2', 'step_3', 'confirm']

# Identify repeated steps:
all_df['next_step'] = all_df.groupby('client_id')['process_step'].shift(-1)

# Correctly check for backtracking
def check_backtrack(row):
    if row['next_step'] in funnel_orders and row['process_step'] in funnel_orders:
        return funnel_orders.index(row['next_step']) < funnel_orders.index(row['process_step'])
    return False

all_df['backtrack'] = all_df.apply(check_backtrack, axis=1)
```

```python
all_df
```

```python
# Focus only on the last valid full path each client takes through the process (ignoring intermediate errors).

unique_steps = all_df[~all_df['backtrack']].groupby(['client_id', 'process_step']).first().reset_index()
```

```python
# Track the successful attempts for each step and see who moved to the last valid consecutive step.
completion_counts = unique_steps.groupby('process_step')['client_id'].nunique()
total_clients_start = unique_steps[unique_steps['process_step'] == 'start']['client_id'].nunique()
completion_rates = completion_counts / total_clients_start
completion_rates
```

```python
# Now the same but diferetiating "Test" and "Control"

valid_steps = all_df[~all_df['backtrack']] # Filter to remove backtracking steps

# Split the data into Test and Control groups
test_group = valid_steps[valid_steps['Variation'] == 'Test']
control_group = valid_steps[valid_steps['Variation'] == 'Control']
```

```python
# Function to calculate completion rates for a group
def calculate_completion_rates(group, funnel_orders):
    # Calculate total unique clients who started
    total_clients_start = group[group['process_step'] == 'start']['client_id'].nunique()

    # Calculate unique clients who completed each step
    completion_counts = group[group['process_step'].isin(funnel_orders)].groupby('process_step')['client_id'].nunique()
    
    # Calculate the completion rates
    completion_rates = completion_counts / total_clients_start
    
    return completion_rates

# Calculate completion rates for both Test and Control
test_completion_rates = calculate_completion_rates(test_group, funnel_orders)
control_completion_rates = calculate_completion_rates(control_group, funnel_orders)

"Test", test_completion_rates, "Control", control_completion_rates
```
